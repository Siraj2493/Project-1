```python
#msn_codes['unit_code'] = list(msn_codes['MSN'])[4]
#df['name'] = df['name'].map(lambda name: name.upper())
msn_codes['unit_code'] = msn_codes['MSN'].map(lambda MSN: list(MSN)[4])
msn_codes['sector_code'] = msn_codes['MSN'].map(lambda MSN: list(MSN)[2] + list(MSN)[3])
msn_codes['type_code'] = msn_codes['MSN'].map(lambda MSN: list(MSN)[0] + list(MSN)[1])
msn_codes
```

```python
sectors = {}
fuels = {}
def get_info(row):
    if 'consumed by the' in row['Description']:
        #print(row['Description'])
        fuels[row['type_code']] = row['Description'].split(' consumed by the ')[0]
        sectors[row['sector_code']] = row['Description'].split(' consumed by the ')[1].split(' sector')[0]
        #fuels[row['sector_code']] = row['Description'].split('consumed by the')[1][:-len(8)]
        #temp_fuels = row['Description'].split('consumed by the')[1]
        #fuels[row['sector_code']] = temp_fuels[:-len(8)]

for row in msn_codes.iterrows():
    get_info(row[1])
    
sectors['TC'] = 'total'
sectors
```

```python
sector_load = {'code' : [], 'sector' : []}
for k, v in sectors.items():
    sector_load['code'].append(k)
    sector_load['sector'].append(v)
    
sector_load
```
