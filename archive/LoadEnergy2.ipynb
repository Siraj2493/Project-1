```python
import pandas as pd
debug_flag = False
```

```python
# constants and converstions
coal_in_btu = 18911
oil_in_btu = 5705
natural_gas_in_btu = 1035
other_gases = 0.001

if debug_flag == True:
    coal_in_btu = 2
    oil_in_btu = 4
    natural_gas_in_btu = 8

def convert_unit(energy_type, value):
    if energy_type == "Coal (Short Tons)":
        return value * coal_in_btu
    elif energy_type == "Petroleum (Barrels)":
        return value * oil_in_btu
    elif energy_type == "Natural Gas (Mcf)":
        return value * natural_gas_in_btu
    else:
        return value

print(convert_unit("Coal (Short Tons)",5))
print(convert_unit("Petroleum (Barrels)",5))
print(convert_unit("Natural Gas (Mcf)",5))
```

```python
# filter "Total Electic Power Industry"
producer_type_sr =  energy_df["TYPE OF PRODUCER"]=="Total Electric Power Industry"
total_df = energy_df[producer_type_sr]
total_df.head()
```

```python
total_df = total_df.merge(states_df, on='Abbreviation')
total_df.rename(columns = {'State_x':'State'}, inplace = True)
total_df.head()
```

```python
# filter "Natural Gas (Mcf)"
if debug_flag == True:
    ref_energy_type = "Coal (Short Tons)"
    ref_energy_type = "Petroleum (Barrels)"
    ref_energy_type = "Natural Gas (Mcf)"
    natural_gas_sr1 =  total_df["ENERGY SOURCE AND UNITS"] == ref_energy_type
    natural_gas_df1 = total_df[natural_gas_sr1]
    natural_gas_df1.head()
```

```python
print(type(total_df))
total_computed_df = total_df.copy()
print(type(total_computed_df))
```

```python
# normalize the unit
total_computed_df['CONSUMPTION'] = total_computed_df.apply(lambda x: convert_unit(x[4], x[5]), axis = 1)
total_computed_df.head()
```

```python
# filter "Natural Gas (Mcf)"
if debug_flag == True:
    natural_gas_sr2 =  total_computed_df["ENERGY SOURCE AND UNITS"] == ref_energy_type
    natural_gas_df2 = total_computed_df[natural_gas_sr2]
    natural_gas_df2.head()
```

```python
# of all energy type
total_computed_gbdf = total_computed_df.groupby(["YEAR", "MONTH", "State"]).sum()
total_final_df = total_computed_gbdf.reset_index()

# capitalize the column names
total_final_df.columns = [x.capitalize() for x in total_final_df.columns]
total_final_df.head()
```

```python
# write out to csv file
total_final_df.to_csv('Resources/energy_data.csv', index=False)
```
